# Cleaning The Data
### Tokenizing
### Removing punctuation
### Stopwords
### Lemmatization

```python
def clean(dataset):
    rev_list=[]
    for i in range(0,len(dataset)):
        para = dataset["essay"][i]
        para_tokens = word_tokenize(para)
    
        punctuation = re.compile(r'[-%&#`~.,;:?/\|_!()0-9]')
        post_punctuation = []
        for words in para_tokens:
            word = punctuation.sub("",words)
            if(len(word))>0:
                post_punctuation.append(word.lower())
    
        stopw = set(stopwords.words('english'))
        filtered_sen=[]
        for w in post_punctuation:
            if w not in stopw:
                filtered_sen.append(w);
                
        lemma_str=[]
        lemmatizer = WordNetLemmatizer()
        for word in filtered_sen:
            lemma_str.append(lemmatizer.lemmatize(word))

        joined_str = " ".join(lemma_str)
        rev_list.append(joined_str)
        
    return rev_list
```

```python
train_filter = clean(train_data)
test_filter = clean(test_data)
```

## Cleaning prompts

```python
def clean(dataset):
    rev_list=[]
    for i in range(0,len(dataset)):
        para = dataset["prompt_question"][i]
        para_tokens = word_tokenize(para)
    
        punctuation = re.compile(r'[-%&#`~.,;:?/\|_!()0-9]')
        post_punctuation = []
        for words in para_tokens:
            word = punctuation.sub("",words)
            if(len(word))>0:
                post_punctuation.append(word.lower())
    
        stopw = set(stopwords.words('english'))
        filtered_sen=[]
        for w in post_punctuation:
            if w not in stopw:
                filtered_sen.append(w);
                
        lemma_str=[]
        lemmatizer = WordNetLemmatizer()
        for word in filtered_sen:
            lemma_str.append(lemmatizer.lemmatize(word))

        joined_str = " ".join(lemma_str)
        rev_list.append(joined_str)
        
    return rev_list
```

```python
prompt_fil = clean(prompts)
prompt_fil
```

## Adding features:  

```python
def addfeatures(prompt_fil,train_filter,train_data):
    total_data = []
    for i in range(0,len(train_filter)):
        data = []
        lex = LexicalRichness(train_data.iloc[i,2])
        score=0
        id = train_data.iloc[i,0]
        prompt = []
        if id==5:
            prompt = prompt_fil[-1]
        else:
            prompt = prompt_fil[train_data.iloc[i,0]-1]
        for word in prompt:
            ans = train_filter[i].find(word)
            if ans!=-1:
                score = score+1
        data.append(score)
        data.append(lex.words)
        data.append(lex.terms)
        data.append(lex.ttr)
        data.append(lex.mtld(threshold=0.72))
        total_data.append(data)
    return total_data
```
